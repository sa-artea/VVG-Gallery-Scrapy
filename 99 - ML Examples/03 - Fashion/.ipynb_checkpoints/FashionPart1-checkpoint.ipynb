```python
#Imports and dependencies 
import numpy as np
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
from time import time

print("Import Success");
```

Datos de entrenamiento: 60 mil imagenes con su correspondiente etiqueta.
Datos de prueba: 10 mil imagenes.

X = Imagenes 
y = Etiquetas

Las imagenes tiene 28 pixeles de ancho por 28 pixeles de alto. 
Se tiene como paramtros de entrada cada uno de los pixeles para un total 784 entradas

### Carga de datos 
Se cargan los datos apartir de la base de datos de keras, que corresponden a los mismos datos de partida para la realizacion de este taller 
Se realiza una funcion para poder verificar la etiqueta y la imagen que corresponden a un elemnto del arreglo. 

```python
def plotImage(y, x):
    print("Number output is : ", y)
    plt.imshow(x, cmap='Greys')

(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()

## Config
pixelW = 28
pixelH = 28
outputs = 10
epochs = 10
regularizationTerm = 0.01

plotImage(y_train[500], x_train[500])

##model = keras.Sequential()
##model.add(keras.layers.Dense(pixelH*pixelW, activation='sigmoid', kernel_regularizer=keras.regularizers.l1(regularizationTerm)))
##model.add(keras.layers.Dense(pixelH, activation='sigmoid', kernel_regularizer=keras.regularizers.l1(regularizationTerm)))
##model.add(keras.layers.Dense(outputs, activation='softmax', kernel_regularizer=keras.regularizers.l1(regularizationTerm)))
```

### Primera aproximación 
Se resolverá el problema usando CNN. para reducir el número de entradas se hacen dos procedimientos


# Normalizacion:

```python
x_train.shape
y_train.shape

x_train = x_train.reshape(x_train.shape[0], 28, 28, 1)
x_test = x_test.reshape(x_test.shape[0], 28, 28, 1)
input_shape = (28, 28, 1)

x_train = x_train.astype('float32')
x_test = x_test.astype('float32')

x_train /= 255
x_test /= 255
print('x_train shape:', x_train.shape)
print('Number of images in x_train', x_train.shape[0])
print('Number of images in x_test', x_test.shape[0])
```

## Estructura de la red

## Capa 1 - Convolucion
Se trata de reducir el número de pixeles al procesar debido a la relación existente entre pixeles continuos. 
La convolución permite mantener la relación entre diferentes partes de la imagen. Este proceso funciona especialmente bien cuando se trata de procesamiento de imágenes. 
Al final el resultado buscar reducir la complejidad de los datos de entrada

Input_shape =  ancho, alto, canal
Canal = 1, si es escala de grises
Kernel_size = ventanad de reduccion, ventana de la convolucion 

## Capa 2 Pooling 
Se suman elige el valor más representativo de los datos o la sumatoria de un cuadrante, para que represente el valor y la relación de ese conjunto de datos.

## Capa 3 Flatten
Es la transformacion de un arreglo de NxN dimensiones en un vector de 1x(NxN), es decir en un vector undimensional 

## Capa 4 Dense
Capa conectada, todo con todo

## Capa 5 Dropout
Ignorar neuronas en el entranamiento. Esto permite tener una probabilidad en cada iteraccion de ignorar un numero de neuronas. 
Teniendo como resultado una reduccion del numero de nodos en cada iteracion. 
En este caso se usa para reducir o evitar el sobreentendimiento de la red. Sobretodo en la corelacion en una capa conectada completamente. 
Al usar una capa de tipo DENSE

## Capa 6 - SOFMAX 
Permite tener certeza de la probabilidad de la salida. Es euler al valor de la salida, sobre la sumatoria de euler al valor de cada salida de la red. 


# Modelo:

```python
model = keras.Sequential()
model.add(keras.layers.Conv2D(28, kernel_size=(3,3), input_shape=input_shape))
model.add(keras.layers.MaxPooling2D(pool_size=(2, 2)))
model.add(keras.layers.Flatten()) 
model.add(keras.layers.Dense(128, activation=tf.nn.relu))
model.add(keras.layers.Dropout(0.2))
model.add(keras.layers.Dense(10,activation=tf.nn.softmax))
```

## ADAM - adaptive moment estimation
Es un metodo de optimizacion estocastico que permite reducir el tiempo de procesamiento haciendo en algunos casos que la red neuronal converja mucho mas rapido. 
Adam es un metodo adaptativo de la tasa de aprendisaje, esta computa diferentes tasas de aprendisaje para cada parametro.Usando una estimacion del primer y segundo
 momento del gradiente para adaptar cada peso de la red neuronal.

## SGD - stochastic gradient descent
Permite minimizar la funcion de error para hacer converger mas rapido la red durante el entrenamiento. 

# Optimizacion - SGD:

```python
print("Start Evaluation")
start_time = time()
model.compile(optimizer='sgd', 
              loss='sparse_categorical_crossentropy', 
              metrics=['accuracy'])
model.fit(x=x_train,y=y_train, epochs=epochs)
elapsed_time = time() - start_time
print("Elapsed time: %.10f seconds." % elapsed_time)
```

# Optimizacion - ADAM:

```python
print("Start Evaluation Network")
start_time = time()
model.compile(optimizer='adam', 
              loss='sparse_categorical_crossentropy', 
              metrics=['accuracy'])
model.fit(x=x_train,y=y_train, epochs=epochs)
elapsed_time = time() - start_time
print("Elapsed time: %.10f seconds." % elapsed_time)
```

## Comparacion tiempo de entrenamiento
* SGD = 277.27 segundos -> 4 minutos 37 segundos
* ADAM = 289.27 segundos -> 4 minutos 49 segundos

# EVALUACION DEL MODELO

```python
model.evaluate(x_test, y_test)
```

# PRUEBA UNITARIA 

```python
plt.imshow(x_test[43].reshape(28, 28),cmap='Greys')
pred = model.predict(x_test[43].reshape(1, 28, 28, 1))
print(pred)
print("El nuemro es igual a ", pred.argmax())
```
